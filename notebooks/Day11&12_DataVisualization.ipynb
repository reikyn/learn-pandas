
### **What is `matplotlib.pyplot`? 🎨**

`matplotlib.pyplot` (often imported as `plt`) is a collection of functions within the Matplotlib library that provides a **MATLAB-like interface** for plotting.

It is designed to make plotting simple and fast, especially for interactive use or quick, simple plots. The key characteristic of `pyplot` is its **stateful interface**, which means:

* **Implicit Tracking:** `pyplot` implicitly keeps track of the "current" Figure and the "current" Axes (the actual plot area).
* **Convenience Functions:** Functions like `plt.plot()`, `plt.title()`, `plt.xlabel()`, and `plt.show()` are wrappers that automatically apply the changes to the current plot without you having to explicitly reference the figure or axes objects.
* **The Object-Oriented (OO) Alternative:** While `pyplot` is easy for quick plots, Matplotlib's recommended approach for complex, customizable visualizations is the **Object-Oriented (OO) style**. In the OO style, you explicitly create and work with `Figure` and `Axes` objects (e.g., `fig, ax = plt.subplots()` and then use methods like `ax.plot()`). The `pyplot` module is still commonly used to create the initial `Figure` and `Axes` objects, even when using the OO approach.

### **The `plot()` Method and its Fundamentals**

The `plot()` function—specifically `matplotlib.pyplot.plot()` or the `Axes.plot()` method—is the primary tool used to create 2D line plots and scatter plots.

#### 1. Basic Call Signatures (Arguments)

The `plot()` function is versatile, but the most fundamental ways to use it involve passing coordinate data and an optional format string:

| Call Signature | Description |
| :--- | :--- |
| `plt.plot(y)` | If you provide **only one sequence of data (`y`)**, Matplotlib assumes these are the **y-values** and automatically generates the x-values as `[0, 1, 2, 3, ...]` (the index of the data). |
| `plt.plot(x, y)` | Plots `y` versus `x` using a **default line style**. |
| `plt.plot(x, y, fmt)` | Plots `y` versus `x` using a specified **format string** (`fmt`) to control color, marker, and line style. |

#### 2. The Format String (`fmt`)

The most common way to quickly customize your plot is with the optional third argument, the **format string (`fmt`)**. It is a shorthand way to specify the **color**, **marker style**, and **line style** of the plot, following the pattern: `fmt = '[marker][line][color]'`.

| Component | Examples |
| :--- | :--- |
| **Color** | `'r'` (red), `'g'` (green), `'b'` (blue), `'k'` (black) |
| **Marker** | `'o'` (circle), `'s'` (square), `'^'` (triangle), `'.'` (point), `''` (no marker) |
| **Line Style** | `'-'` (solid line), `'--'` (dashed line), `'-.'` (dash-dot), `':'` (dotted), `''` (no line) |

**Example:** `plt.plot(x, y, 'ro--')` will produce a plot with **r**ed, **o**utlined circle markers, connected by a **--** dashed line.

#### 3. Key Keyword Parameters (`kwargs`)

Beyond the position arguments (`x`, `y`, `fmt`), you can use keyword arguments (`kwargs`) to gain finer control over the line's appearance:

| Parameter | Purpose | Example |
| :--- | :--- | :--- |
| **`color` or `c`** | Sets the line color (overrides `fmt` color). | `color='green'` |
| **`linestyle` or `ls`** | Sets the line style (overrides `fmt` line style). | `linestyle=':'` (dotted) |
| **`linewidth` or `lw`** | Sets the thickness of the line. | `linewidth=2.5` |
| **`marker`** | Sets the shape of the marker at each data point. | `marker='^'` |
| **`markersize` or `ms`** | Sets the size of the marker. | `markersize=8` |
| **`label`** | A string used for the legend. | `label='Data Series 1'` |
| **`data`** | Allows you to pass a Pandas DataFrame or other structured object and reference columns by string in `x` and `y`. | `plt.plot('Date', 'Close', data=df)` |

````python
# Recommended approach
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# Flatten axes for easy iteration if needed
axes = axes.flatten()

numerics = df.select_dtypes(include='number').columns.tolist()

for i, ax in enumerate(axes):
    # ax.plot(np.random.randn(100).cumsum())
    ax.plot(df[numerics[i]], ls='--', marker='o')
    ax.set_title(f'Subplot {i+1}')
    ax.grid(True)

plt.tight_layout()
plt.show()
````


### **Matplotlib Anatomy: Figure and Axes**

The code snippet you saw (`fig, ax = plt.subplots(...)`) uses the recommended **Object-Oriented (OO) interface** of Matplotlib. This interface is built on a simple hierarchy:

| Component | Analogy | Description |
| :--- | :--- | :--- |
| **Figure (`fig`)** | The **Canvas** or **Picture Frame** | The outermost container. It holds everything: the plot area, titles, labels, and multiple subplots (if you have them). You can control the size of the overall image here. |
| **Axes (`ax`)** | The **Actual Plot Area** or **Graph Paper** | The region where the data is actually plotted. This is where you call methods like `.plot()`, `.bar()`, `.set_xlabel()`, and `.set_title()`. A Figure can contain multiple Axes. |



### **`plt.figure()`: The Canvas Creator**

The `plt.figure()` function is the most fundamental way to create a new **Figure** object.

| Purpose | Creates an empty Figure, or retrieves the current one. |
| :--- | :--- |
| **Returns** | A single **`Figure`** object. |
| **Key Methods/Parameters** | **`plt.figure(figsize=(w, h), dpi=d)`** |
| `figsize` | A tuple `(width, height)` in **inches**. E.g., `figsize=(8, 6)`. |
| `dpi` | The **dots per inch** (resolution) of the figure. Higher DPI means sharper image, but a larger file size. |

#### Why Use `plt.figure()`?

You use this when you need an empty canvas to explicitly add an Axes object to later (e.g., using `fig.add_axes()`), or as the first step before calling stateful `pyplot` commands.




### **`plt.subplots()`: The All-in-One Utility**

This is the most common and recommended way to start plotting, as it instantly gives you both the canvas and the plot area.

| Purpose | Creates a Figure and one or more Axes (subplots) simultaneously. |
| :--- | :--- |
| **Returns** | A tuple: **`(Figure, Axes or Array of Axes)`**. This is why you saw `fig, ax = ...` |
| **Key Methods/Parameters** | **`plt.subplots(nrows=n, ncols=m, figsize=(w, h))`** |
| `nrows` | The number of rows in the subplot grid (default is $1$). |
| `ncols` | The number of columns in the subplot grid (default is $1$). |
| `figsize` | Same as `plt.figure()`, sets the size of the overall canvas. |

#### Why Use `plt.subplots()`?

It simplifies the code and immediately puts you into the **Object-Oriented (OO) plotting style**, which is generally preferred because:

1.  **Clarity:** You explicitly modify the plot object (`ax` or `axs`), making your code less prone to errors in complex multi-plot figures.
2.  **Flexibility:** It's the standard way to create multiple plots in one window (subplots).



#### Understanding the Return Values:

  * **For a single plot (`nrows=1, ncols=1`):**
    ```python
    fig, ax = plt.subplots(figsize=(8, 6))
    # 'ax' is a single Axes object.
    ax.bar(...)
    ```
  * **For multiple plots (e.g., $2$ rows, $2$ columns):**
    ```python
    fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))
    # 'axs' is a 2D NumPy array of Axes objects.
    axs[0, 0].plot(...) # Top-left plot
    axs[1, 1].hist(...)  # Bottom-right plot
    ```



### **Key `Axes` Object Methods (The OO Style)**

Once you have the `ax` object, you use its methods to draw and customize the plot. This is the OO approach.

| Pyplot (`plt.`) Style | Axes (`ax.`) Style (Recommended) | Purpose |
| :--- | :--- | :--- |
| `plt.plot(x, y)` | **`ax.plot(x, y)`** | Draws a line/scatter plot. |
| `plt.title('Title')` | **`ax.set_title('Title')`** | Sets the plot title. |
| `plt.xlabel('X-Label')` | **`ax.set_xlabel('X-Label')`** | Sets the X-axis label. |
| `plt.ylabel('Y-Label')` | **`ax.set_ylabel('Y-Label')`** | Sets the Y-axis label. |
| `plt.legend()` | **`ax.legend()`** | Displays the legend. |
| `plt.xticks(...)` | **`ax.set_xticks(...)`** | Sets the tick locations. |

The power of the **Object-Oriented (OO) style** is that once you have your $\texttt{ax}$ object (from $\texttt{fig, ax = plt.subplots()}$), you simply replace $\texttt{plt.}$ with $\texttt{ax.}$ for all these plotting methods (e.g., $\texttt{ax.plot()}$, $\texttt{ax.bar()}$).

````python
# Let's see the average Anxiety Level and Sleep Hours of Student based on their Placement Status
# 1. create the data
avg_anx_slp = df.groupby('Placement_Status')[['Anxiety_Level', 'Sleep_Hours']].mean()
# avg_anx_slp.plot(kind='bar') # Using pandas

# 2. Prepare the Data
statuses = avg_anx_slp.index.tolist()
anxiety = avg_anx_slp['Anxiety_Level'].values
sleep = avg_anx_slp['Sleep_Hours'].values

# 3. Define bar parameters
x = np.arange(len(statuses))  # the label locations (0, 1, 2, ...)
width = 0.35  # the width of the bars

# 4. Create the plot using plt.subplots() and plt.bar()
fig, ax = plt.subplots(figsize=(10, 6))

# Plot Anxiety_Level bars (shifted left by half a bar width)
rects1 = ax.bar(x - width/2, anxiety, width, label='Average Anxiety Level', color='blue')
# Plot Sleep_Hours bars (shifted right by half a bar width)
rects2 = ax.bar(x + width/2, sleep, width, label='Average Sleep Hours', color='orange')


# 5. Customize the plot
ax.set_ylabel('Average Score / Hours')
ax.set_title('Average Anxiety Level and Sleep Hours by Placement Status')
ax.set_xticks(x)
ax.set_xticklabels(statuses)
ax.legend(loc='upper right') # Add a legend
ax.grid(axis='y', linestyle='--', alpha=0.7) # Add a horizontal grid
plt.xticks(rotation=0) # Keep x-axis labels horizontal
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show() # Display the plot (in a real environment)
````

````python
# Create a Scatter Plot to explore the relationship between Anxiety_Level and Sleep_Hours. We'll also use the c (color) parameter to add a third dimension: Confidence_Level. This helps us see if confidence influences the Anxiety/Sleep relationship.

fig, ax = plt.subplots(figsize=(8,6))
scatter = ax.scatter(df['Anxiety_Level'], df['Sleep_Hours'], c=df['Confidence_Level'], s=100, cmap='rainbow', edgecolors='black')

ax.set_title('Anxiety vs. Sleep, Colored by Confidence Level')
ax.set_xlabel('Anxiety Level')
# ax.set_y/label('Sleep Hours')
ax.grid(True, linestyle='--')

cbar = fig.colorbar(scatter, ax=ax)
cbar.set_label('Confidence Level (Score)')
````

````python
# Your Task: Create a Pie Chart to visualize the distribution of students across the different Branch categories.

# 1. Prepare the Data:
branch_count = df['Branch'].value_counts()
sizes = branch_count.values
labels = branch_count.index.tolist()

# 2. Create Figure and Axes
fig, ax = plt.subplots(figsize=(6, 6)) # Pie charts look best in a square figure

# 3. Use the .pie() method
ax.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',      # Format to display percentage (e.g., 25.0%)
    startangle=90,          # Start the first slice at the top
    shadow=True,
    explode=(0.1, 0, 0, 0, 0, 0)  # Explode the largest slice (CSE) for emphasis
)

# 4. Customization
ax.set_title('Students Branch Distribution')
ax.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()

# Save the plot
# plt.savefig('branch_distribution_pie_chart.png')
````

## **The Matplotlib & Seaborn Ecosystem 📊**

### **1. Introduction: Seaborn Analogy and Purpose**

| Component | Analogy | Description |
| :--- | :--- | :--- |
| **Matplotlib** | The **Canvas** or **"Engine"** | The foundation. It provides the low-level functions for drawing lines, setting colors, and manipulating every element on the plot. It offers **maximum control**. |
| **Seaborn** | The **Painter** or **"Template Designer"** | Built on top of Matplotlib, Seaborn is a high-level library. It uses sensible defaults and complex algorithms to automatically create beautiful, statistically informative graphics with minimal code. It offers **maximum convenience**. |

**Purpose:** Seaborn is explicitly designed for **statistical visualization**. It provides integrated functionality for plotting relationships between multiple variables (like using `hue` or `size`), visualizing distributions, and working with complex data structures like correlation matrices.



### **2. Comparison: Matplotlib🎨 vs. Seaborn🌊**

The industry often uses **both** libraries in conjunction, capitalizing on their strengths: Seaborn for rapid, beautiful statistical exploration, and Matplotlib for fine-tuning, customization, and managing the Figure and Axes objects.

| Feature | Matplotlib (The Foundation) | Seaborn (The Statistical Layer) |
| :--- | :--- | :--- |
| **Level** | Low-level API (The `ax.plot()` and `ax.set_...` methods). | High-level API (The `sns.scatterplot()` and `sns.heatmap()` functions). |
| **Data Focus** | General plotting (lines, bars, simple shapes). | Statistical relationships and distributions (regressions, comparisons). |
| **Default Aesthetics** | Simple, sometimes dated appearance. Requires manual styling. | Modern, appealing default styles and color palettes. |
| **Handling Data** | Requires data preparation (e.g., grouping, sorting) before plotting. | Often takes the entire **Pandas DataFrame** as a primary argument, handling grouping and statistical aggregation internally. |
| **Industry Use** | For **final production** plots where complete control over fonts, layout, and annotations is needed; also used to define the `fig`, `ax` structure. | For **Exploratory Data Analysis (EDA)**, quick statistical insights, and high-quality report graphics. |




### **3. Dedicated Seaborn Plotting Methods**

Seaborn functions are often categorized by the type of relationship they visualize. They typically accept the DataFrame itself (`data=df`) and column names as strings (`x='col_name'`, `y='col_name'`).

| Method | Purpose | Matplotlib Equivalent/Relationship |
| :--- | :--- | :--- |
| **`sns.heatmap()`** | Visualize matrix data (like correlation) as a color-coded map. | Uses Matplotlib's `ax.imshow()` under the hood. |
| **`sns.scatterplot()`** | Advanced scatter plots with easy integration of color (`hue`) and size based on other variables. | More sophisticated version of `ax.scatter()`. |
| **`sns.kdeplot()`** | Visualize the distribution of a single or two variables using **Kernel Density Estimation**. | Related to `ax.hist()` but shows a smoothed probability curve. |
| **`sns.boxplot()`** | Similar to `ax.boxplot()` but with cleaner defaults and easier grouping using the `x` and `y` parameters. | Same as `ax.boxplot()` but with simplified syntax. |
| **`sns.pairplot()`** | Creates a grid of scatter plots for every pair of variables in a DataFrame. **Essential for EDA.** | Requires a complex combination of `plt.subplots()` and `ax.scatter()` calls in pure Matplotlib. |
| **`sns.lineplot()`** | Statistical line plots, often showing central tendency and confidence intervals. | Related to `ax.plot()` but optimized for time series or continuous data with multiple observations per point. |




### **4. Fundamental Seaborn Parameters**

Seaborn functions reduce the amount of code needed by integrating key parameters that Matplotlib handles separately:

| Parameter | Purpose | Example |
| :--- | :--- | :--- |
| **`data`** | The DataFrame object containing the data. | $\texttt{data=df}$ |
| **`x` / `y`** | The column names for the X and Y axes (passed as strings). | `x='Anxiety_Level'`, `y='Sleep\_Hours'` |
| **`hue`** | The column name used to group and color the data points/bars. | `hue='Placement_Status'` |
| **`size`** | The column name used to set the size of markers (in scatter plots). | `size='Confidence\_Level'` |
| **`cmap`** | The color palette to use for the visualization (same as Matplotlib). | `cmap='viridis'` |


#### Practice 1: The Heatmap (`sns.heatmap()`)

**Task:** Create a Heatmap to visualize the **correlation matrix** between all numerical columns in your ${df}$.

  * We'll use `.corr()` to calculate the relationships.
  * We'll use `sns.heatmap()` to plot the matrix.


````python
# 1. Prepare the data:
numeric_cols = df.select_dtypes(include='number')
corelation = numeric_cols.corr()

# 2. Create Figure and Axes
fig, ax = plt.subplots(figsize=(8, 6))

# 3. Create heatmap using seaborn's .heatmap() method
sns.heatmap(
    corelation,
    ax=ax,              # Crucially, tell Seaborn which Matplotlib Axes to draw on
    annot=True,         # Display the correlation values on the map
    linewidths=.5,      # Colormap (red for negative, blue for positive)
    cmap='coolwarm',    # Format the annotations to two decimal places
    fmt='.2f',           # Add lines between cells for clarity
    cbar_kws={'label': 'Correlation Coefficient'}
)

# 4. Customization
ax.set_title('Correlation Matrix of Student Factors')
plt.tight_layout()

# Save the plot
# plt.savefig('correlation_heatmap.png')
````

#### Practice 2: Voilin Plot

**Task:** Create a Violin Plot to visualize and compare the distribution of *Anxiety_Level* across the different *Branch* categories.

  * We'll use `sns.violinplot()` to plot violin plot.
  * We'll set the `inner='quartile'` parameter to display the median and interquartile range inside the plot, just like a traditional Box Plot.

**Key Parameters** for `sns.violinplot()`:
  * **x**: The categorical column (e.g., ’Placement_Status’).

  * **y**: The numerical column whose distribution you want to show (e.g., ’Anxiety_Level’).

  * **data**: The DataFrame containing the columns.

  * **palette**: The color palette to use for the categories.

  * **inner**: What to draw inside the violins (e.g., ’quartile’ for lines at Q1, median, and Q3).

````python
# 1. Create Figure and Axes:
fig, ax = plt.subplots(figsize=(8, 6))

# 2. Create violin plot
sns.violinplot(
    x='Branch',
    y='Anxiety_Level',
    data=df,
    ax=ax,
    hue='Branch',
    palette='Spectral',
    legend=False,
    inner='quartile',
    linewidth=1.5
)

# 3. Customization
ax.set_title('Anxiety Level Distribution by Branch (Violin Plot)')
ax.set_xlabel('Branch')
ax.set_ylabel('Anxiety Level Score')
ax.grid(axis='y', linestyle='--', alpha=0.8)
plt.tight_layout()

# Save the plot
# plt.savefig('anxiety_level_violin_plot.png')
````

#### Practice 3: The Categorical Scatter Plot (Swarm Plot: sns.swarmplot)

**Task**: Create a Swarm Plot to visualize the individual Sleep_Hours of students across the different Branch categories.

**Key Parameters** for sns.swarmplot:

* **x**: The categorical column (e.g., ’Branch’).

* **y**: The numerical column whose individual points you want to plot (e.g., ’Sleep_Hours’).

* **data**: The DataFrame containing the columns.

* **hue**: An optional categorical column to color the points by (e.g., ’Placement_Status’).

* **size**: Controls the size of the markers (points).

````python
# 1. Create Figure and Axes
fig, ax = plt.subplots(figsize=(10, 6))

# 2. Use the seaborn.swarmplot() method
# We will use 'Placement_Status' for the 'hue' parameter to add a fourth dimension.
sns.swarmplot(
    x='Branch',
    y='Sleep_Hours',
    hue='Placement_Status',
    data=df,
    ax=ax,
    size=7,
    palette='Set2',
    dodge=True # Separate points based on 'hue' variable
)

# 3. Customization
ax.set_title('Sleep Hours by Branch, Colored by Placement Status (Swarm Plot)')
ax.set_xlabel('Student Branch')
ax.set_ylabel('Sleep Hours')
ax.grid(axis='y', linestyle='--', alpha=0.5)
plt.legend(title='Placement Status', loc='upper right')
plt.tight_layout()

# Save the plot
# plt.savefig('sleep_hours_swarm_plot.png')
````

#### Practice 4: Pair Plot

**Task**: Create a Pair Plot for all the numerical columns in the dataset, coloring the points by Placement_Status to see how placement status influences the relationships.

**Key Parameters** for sns.pairplot
* **data**: The DataFrame containing the columns.

* **vars**: A list of column names to include in the grid (if you don't want all numerical columns).

* **hue**: The categorical column used to color all the plots (e.g., ’Placement_Status’).

* **diag_kind**: The type of plot to draw on the diagonal (e.g., ’hist’ or ’kde’).

* **kind**: The type of plot to draw on the off-diagonals (e.g., ’scatter’ or ’reg’).
