## **Introduction üé®**

**Data visualization** is the process of presenting data in a visual context, such as a chart, graph, or map. In today‚Äôs data-driven world, it's a critical skill that translates complex datasets into clear, digestible, and actionable insights.

### **The Need for Visualization in AI/ML**

In the fields of Artificial Intelligence and Machine Learning, visualization is not just for pretty pictures; it's a fundamental part of the workflow.

  * **Exploratory Data Analysis (EDA):** Before building a model, visualizations help us understand the data's distribution, identify outliers, spot trends, and uncover relationships between variables. This is often the first step in any data science project.
  * **Model Performance Evaluation:** Visualizations are used to assess how well a model is performing. For example, a confusion matrix or a ROC curve plot can quickly show a model's strengths and weaknesses.
  * **Communicating Insights:** The best AI/ML models are useless if their results can't be communicated effectively. Visualizations help data scientists and engineers explain their findings to non-technical stakeholders, ensuring the insights drive informed business decisions.

### **Visualization in Python: A Quick Look at the Tools**

We will explore three key libraries for data visualization in Python:

  * **Pandas:** As a data analysis tool, Pandas has built-in plotting capabilities. It‚Äôs perfect for creating quick, basic plots directly from a DataFrame, making it incredibly convenient for rapid exploration.
  * **Seaborn:** Built on top of Matplotlib, Seaborn is a high-level library designed for creating beautiful and informative statistical graphics. It has a more user-friendly syntax and comes with pre-defined styles and color palettes.
  * **Matplotlib:** The foundation of plotting in Python, Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations. While it can be more verbose, it offers complete control over every element of a plot.

## **Part 1: Data Visualization with Pandas** üêº

Pandas makes plotting straightforward by integrating with Matplotlib. You can create a plot directly from a DataFrame or a Series by simply calling the `.plot()` method. This is perfect for quick data checks and initial visualizations.

The `.plot()` method has a number of parameters to specify the type of plot you want to create. Here are some of the most common ones you'll use:

  * **`kind`**: The type of plot. Examples include `'line'`, `'bar'`, `'hist'`, `'box'`, and `'scatter'`.
  * **`x`** & **`y`**: The column names to use for the x-axis and y-axis.
  * **`title`**: A string for the plot's title.
  * **`figsize`**: A tuple to set the figure size (width, height).


### **Pandas Plot Types Cheat Sheet**

This table summarizes the most common Pandas plotting functions, their purposes, and basic syntax.

| Plot Type | Purpose (Why & When to Use) | Basic Syntax |
|-----------|-----------------------------|--------------|
| **line**  | Show trends or changes over continuous data (e.g., time series). Useful when order matters. | `df.plot(x='col_x', y='col_y', kind='line')` |
| **bar**   | Compare quantities across categories. Height shows magnitude. | `df.plot(x='category_col', y='value_col', kind='bar')` |
| **barh**  | Same as bar chart, but horizontal. Good for long category names. | `df.plot(x='category_col', y='value_col', kind='barh')` |
| **hist**  | Show distribution of a numeric variable by grouping into bins. | `df['col'].plot(kind='hist', bins=10)` |
| **box**   | Show distribution summary (median, quartiles, outliers). Good for spotting skewness and outliers. | `df.plot(y='col', kind='box')` |
| **area**  | Similar to line plot but filled area under curve. Good for cumulative data visualization. | `df.plot(x='col_x', y='col_y', kind='area', alpha=0.5)` |
| **pie**   | Show proportion of categories as slices of a circle. Works best with few categories. | `df['col'].value_counts().plot(kind='pie', autopct='%1.1f%%')` |
| **scatter** | Show relationship between two numeric variables. Useful for correlation analysis. (DataFrame only). | `df.plot(x='col_x', y='col_y', kind='scatter')` |
| **hexbin** | Show density of points in 2D with hexagonal bins. Great for large scatterplots. (DataFrame only). | `df.plot(x='col_x', y='col_y', kind='hexbin', gridsize=25)` |


Now, let's dive into some practical examples.


```python
import pandas as pd
import numpy as np

# Create the DataFrame
data = pd.read_csv(r'..\data\placement_pressure_pulse.csv')

df = pd.DataFrame(data)

# Add a numerical category for scatter plots
df['Support_System_Score'] = df['Branch'].apply(
    lambda x: np.random.randint(70, 100) if x in ['CSE', 'IT'] else np.random.randint(40, 70)
)

df
```

```python
# Total applications sent by each branch (by line plot)

df.groupby('Branch', sort=False)['Applications_Sent'].sum().plot(ylabel='Applications Sent', xlabel='Branch', title='Monthly Applications Sent', figsize=(8, 5), marker='o')
```

```python
# Bar Plot: Shows the average Confidence_Level by Branch
df.groupby('Branch')['Confidence_Level'].mean().plot(kind='bar', xlabel='Month', title='Avg Confidence Level')
```

```python
# Barh Plot: Shows the average Confidence_Level by Branch
df.groupby('Branch')['Confidence_Level'].mean().plot(kind='barh', xlabel='Confidence Level', title='Avg Confidence Level')
```

```python
# Histogram plotting: Show distribution of a numeric variable by grouping into bins.
df[['Anxiety_Level', 'Confidence_Level']].plot(kind='hist', bins=20, stacked=True, edgecolor='black', figsize=(9,5))
```

```python
# Box plotting: Show distribution summary (median, quartiles, outliers). Good for spotting skewness and outliers.
df.plot(y='Interview_Calls', ylabel='Interview Calls', kind='box', title='boxplot')
```

```python
# area: Similar to line plot but filled area under curve. Good for cumulative data visualization.
# df.plot(x='Placement_Status', y='Interview_Calls', kind='area', title='Area Plot')
# average number of 'Interview_Calls' for each 'Placement_Status' category.
df.groupby('Placement_Status')['Interview_Calls'].mean().plot(kind='area')
df.groupby('Placement_Status')['Interview_Calls'].mean()
```

```python
# pie: Show proportion of categories as slices of a circle. Works best with few categories.
df['Placement_Status'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, wedgeprops={'edgecolor': 'black'}, fontsize=12, figsize=(5,5))
```

```python
# scatter: Show relationship between two numeric variables. Useful for correlation analysis. (DataFrame only).
df.plot(kind='scatter', x='Interview_Calls', y='Applications_Sent', title='Interview Calls vs. Applications Sent', figsize=(8,6), color='darkblue', alpha=0.7)
```

```python
# hexbin: Show density of points in 2D with hexagonal bins. Great for large scatterplots. (DataFrame only).
df.plot(
    kind='hexbin',
    y='Applications_Sent',
    x='Interview_Calls',
    gridsize=10, # A small gridsize for this small dataset to show the concept
    cmap='Blues', # Use a color map to show density
    title='Density of Interview Calls vs. Applications Sent',
    figsize=(8, 6)
)
```
