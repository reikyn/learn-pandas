## **üèÜAdvanced-Level Questions**

```python
# 1. Compare the sales trend before and after 2020:
# Compute the average sales from 2010‚Äì2019 vs 2020‚Äì2024.
# Which period had higher sales?
avg_early = df.loc[df['Year'].between(2010, 2019), 'Sales_Volume'].mean()
avg_recent = df.loc[df['Year'].between(2020, 2024), 'Sales_Volume'].mean()

if avg_recent > avg_early:
    print("2020‚Äì2024 has higher sales")
else:
    print("2010‚Äì2019 has higher sales")
```

```python
# 2. Group the data by Region and find:
# The year with maximum sales for each region.
# The mean sales for each region.

df_grouped = df.groupby('Region')
df_grouped['Sales_Volume'].mean()
df_grouped['Sales_Volume'].idxmax()   # gives row index of max sales
```

```python
# 3. Sort all years by sales, then filter the top 3 years with the highest sales per region.
df.sort_values(by=['Region','Sales_Volume'], ascending=[True,False]).groupby('Region').head(3)
```

```python
# 4. Create a custom function sales_summary(df) that:
# Returns min, max, mean, median, std, and IQR of the sales column.
# Apply it to the BMW dataset.
def sales_summary(df):
    q25 = df['Sales_Volume'].quantile(0.25)
    q75 = df['Sales_Volume'].quantile(0.75)
    iqr = q75 - q25
    
    stats = df['Sales_Volume'].agg(['min','max','mean','median','std'])
    stats['IQR'] = iqr
    
    return stats

sales_summary(df)
```
