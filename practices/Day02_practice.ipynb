## **üéØIntermediate-Level Questions**

```python
# Get all rows where Age is greater than 21.
df.loc[df['Age'] > 21]
```

```python
# Get all students whose Country is either USA or UK.
df[df['Country'].isin(['USA', 'UK'])]
df[(df['Country'] == 'USA') | (df['Country'] == 'UK')]
```

```python
# Select rows where Age is between 18 and 21.
df.query('18 <= Age <= 21')
# df[df['Age'].between(18, 21)]
```

```python
# Get only the Acedemic Score and Age of students older than 20.
df.loc[df['Age'] > 20, ['Affects_Academic_Performance', 'Age']]
```

***

## **üèÜAdvanced-Level Questions**

```python
# Using multiple conditions: get students where Age > 20 and Country == "South Korea".
df.query('Age > 20 and Country == "South Korea"')
```

```python
# Select all rows except those from Canada.
# df[~df['Country'].eq('Canada')]
df[~df['Country'].isin(['Canada'])]
```

```python
# Use .query() to get students with Age < 22 or Country == "India".
df.query('Age < 22 or Country == "India"')
```

```python
# From the dataset, create a new filtered DataFrame of students who are from "South Korea", Average daily usage is in between 3 to 5 and
# have Age between 19‚Äì23, selecting only Student ID, relationship status, and Gender
filtered = df.query('Country == "South Korea" and 3 <= Avg_Daily_Usage_Hours <= 5 and 19 <= Age <= 23')
filtered[['Student_ID', 'Gender', 'Relationship_Status']]
```
