```python
# 10. Create a function clean_column(df, col_name) that:
# Drops rows where col_name is missing, Removes duplicates, Returns the cleaned column as a Series.
def clean_column(df, col_name):
    # return df[col_name].dropna().drop_duplicates()
    return df.dropna(subset=col_name).drop_duplicates(subset=col_name)[col_name]
clean_column(df, 'Rating')
```

```python
# 11. Design your own cleaning pipeline:
# Apply it on the Netflix dataset and explain why each step improves data quality.
# Step 1: Drop columns with >30% missing.
thresh_value = 0.7 * len(df)
df2 = df.dropna(axis=1, thresh=thresh_value).copy()
df2
# Step 2: Fill categorical columns with mode, numeric with median.
fill_values = {col: df2[col].mode()[0] for col in df2.columns if df2[col].isna().sum()}
df2.fillna(value=fill_values, inplace=True)
# Step 3: Remove duplicates.
df2.drop_duplicates(inplace=True)
# Step 4: Reset the index.
df2.reset_index(drop=True, inplace=True)

df2

# step1 Ensures that we have only sufficient data, avoiding noise from mostly empty columns
# step2 Efficiently handle the missing data in a systemetic way, reducing information loss while keeping consistency.
# step3 Removes duplicate rows so analysis isn't biased by repeaded data.
# step4 Resets index so rows are properly numbered after dropping and deduplicating
```

```python
# Conditional Filling:
# For rows where Type == "TV Show" and Director is missing, fill it with "TV Unknown".
# For rows where Type == "Movie" and Director is missing, fill it with "Movie Unknown".
tv = df.copy()
tv.loc[(df['Category'] == 'TV Show') & (df['Director'].isna()), 'Director'] = 'TV Unknown'
tv.loc[(df['Category'] == 'Movie') & (df['Director'].isna()), 'Director'] = 'Movie Unknown'
tv
```

```python
# Custom Imputation Function
# Write a function that:
# Accepts a DataFrame and a list of column names,
# For each numeric column ‚Üí fill missing with median,
# For each categorical column ‚Üí fill missing with mode,
# Returns the cleaned DataFrame.
# Test it on the Netflix dataset.
def clean_columns(df, cols):
    dfc = df.copy()
    for col in cols:
        if dfc[col].dtype == object:
            # dfc[col] = dfc[col].fillna(dfc[col].mode()[0])
            dfc.fillna({col: dfc[col].mode()[0]}, inplace=True)
        else:
            # dfc[col] = dfc[col].fillna(dfc[col].median())
            dfc.fillna({col: dfc[col].median()}, inplace=True)
    return dfc
new_dff = clean_columns(df, ['Show_Id', 'Title','Director', 'Rating'])
new_dff
```
